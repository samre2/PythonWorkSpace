```python
price = 1000  # Global variable


def get_car_price(name, model):
    # This tells Python we want to use and modify the global 'price' variable,
    # instead of creating a new local one.
    global price

    usage_years = 3  # Local variable (only exists inside this function)

    # Let's modify the global price inside the function to see it change
    price = 12000

    print(f"Car: {name} {model}")
    print(f"Usage Years (Local): {usage_years}")
    print(f"Price inside function (Global modified): {price}")


# --- Testing the code ---

# 1. Print global price before running the function
print(f"Global price before function: {price}")
print("-" * 30)

# 2. Call the function with the required arguments (name and model)
get_car_price("Tesla", "Model 3")
print("-" * 30)

# 3. Print global price after the function ran
# You will see it actually changed because we used the 'global' keyword!
print(f"Global price after function: {price}")
```
